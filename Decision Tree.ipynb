## Create the Decision Tree classifier and visualize it graphically.

# Prediction using Decision Tree Algorithm

### Importing the libraries

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import tree
```

### Loading Data

```python
iris = load_iris()
```

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head()
```

```python
df.tail()
```

```python
X, y = iris.data, iris.target
```

### Decision Tree

```python
from sklearn import tree
from IPython.display import Image
import pydotplus
```

```python
clftree=tree.DecisionTreeClassifier(min_samples_leaf=10,max_depth=6)
clftree.fit(X,y)
```

```python
dot_data=tree.export_graphviz(clftree,out_file=None,rounded=True,filled=True,special_characters=True,feature_names=iris.feature_names)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
